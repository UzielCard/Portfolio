### Selección de variables de interes

#### Descripción de las variables seleccionadas

* folioviv: Identificador único de la vivienda, se construye por dígitos de la entidad federativa y otros datos de su posición y se hace al realizar la muestra
* foliohog: Identificador único del hogar. El 1 identifica al hogar principal y del 2 al 5 hogares adicionales.
* clave: El identificador de cada producto adquirido
* tipo_gasto: Tipo de gasto, monetario o no monetario. Se divide en 7 tipos
* lugar_comp: Lugar donde se compraron o consumieron los alimentos, bebidas y/o tabaco los integrantes del hogar
* gasto_tri: Gastos trimestralizados
* cantidad: La cantidad del producto que se counsume. Son litros para las bebidas y kilos para los alimentos

#### Crear dataset solo con los datos de interes

````python
# Query para seleccionar las variables de interes, que devuelve valores unicamente de los productos que queremos analizar y los tipos de gastos que son monetarios
query = """
SELECT
  folioviv,
  foliohog,
  clave,
  tipo_gasto,
  lugar_comp,
  gasto_tri,
  cantidad
FROM gasto_hogar
WHERE clave IN ('A220','A215', 'A209', 'A106', 'A230', 'A233', 'A235')
  AND tipo_gasto IN ('G1');
"""

gasto_hogar_interes = spark.sql(query) # Corre el query, utiliza el método sql() con el query dentro del parentesis para realizar la accion
````

### Join de bases

````python
gasto_hogar_reshape.createOrReplaceTempView('gh_reshape')  # Añadimos nuestro dataset de gasto_hogar_reshape al catálogo como gh_reshape
````

````python
#Generamos un query que una los datasets y seleccione las variables que nos interesa y cree nuevas columnas con calculos. Hace la union cuando folioviv y foliohog son iguales en ambos datasets.
query="""
SELECT
  gh.folioviv,
  gh.foliohog,
  CONCAT('MX',SUBSTRING(ubica_geo, 1, 2)) AS estado,
  A220,
  A215,
  A106,
  A209,
  A230,
  A233,
  A235,
  tam_loc,
  est_socio,
  est_dis,
  CAST(factor AS integer),
  clase_hog,
  CAST(tot_integ AS integer),
  (factor*A220) AS factor_gasto_refresco,
  (factor*A215) AS factor_gasto_agua,
  (factor*A106) AS factor_gasto_papas,
  (factor*A209) AS factor_gasto_golosinas,
  (factor*A230) AS factor_gasto_ron,
  (factor*A233) AS factor_gasto_tequila,
  (factor*A235) AS factor_gasto_vodka
FROM gh_reshape AS gh
  INNER JOIN concentrado_hogar AS ch
    ON gh.folioviv = ch.folioviv
      AND gh.foliohog = ch.foliohog;
"""

consolidado = spark.sql(query) # Corre el query
````

````python
consolidado.show() # Muestra la nueva vista creada por nuestro query
````

# Análisis Exploratorio de la base de datos de la ENIGH

## Estadísticas básicas

### Promedios

````python
#Generamos un query para obtener el promedio de los productos de interes teniendo en cuenta el factor de expansión
query="""
SELECT
  SUM(factor_gasto_refresco)/SUM(factor) AS media_refresco,
  SUM(factor_gasto_agua)/SUM(factor) AS media_agua,
  SUM(factor_gasto_papas)/SUM(factor) AS media_papas,
  SUM(factor_gasto_golosinas)/SUM(factor) AS media_golosinas,
  SUM(factor_gasto_ron)/SUM(factor) AS media_ron,
  SUM(factor_gasto_tequila)/SUM(factor) AS media_tequila,
  SUM(factor_gasto_vodka)/SUM(factor) AS media_vodka
FROM consolidado_gh;
"""

medias = spark.sql(query) # Corre el query
````

````python
medias.show()
````

### Varianzas

````python
##Generamos un query para obtener las varianzas de los productos de interes teniendo en cuenta el factor de expansión
query="""
SELECT
  SUM(factor*(POWER((A220-media_refresco),2)))/SUM(factor) AS var_refresco,
  SUM(factor*(POWER((A215-media_agua),2)))/SUM(factor) AS var_agua,
  SUM(factor*(POWER((A106-media_papas),2)))/SUM(factor) AS var_papas,
  SUM(factor*(POWER((A209-media_golosinas),2)))/SUM(factor) AS var_golosinas,
  SUM(factor*(POWER((A230-media_ron),2)))/SUM(factor) AS var_ron,
  SUM(factor*(POWER((A233-media_tequila),2)))/SUM(factor) AS var_tequila,
  SUM(factor*(POWER((A235-media_vodka),2)))/SUM(factor) AS var_vodka
FROM consolidado_gh
JOIN (
  SELECT
    SUM(factor_gasto_refresco)/SUM(factor) AS media_refresco,
    SUM(factor_gasto_agua)/SUM(factor) AS media_agua,
    SUM(factor_gasto_papas)/SUM(factor) AS media_papas,
    SUM(factor_gasto_golosinas)/SUM(factor) AS media_golosinas,
    SUM(factor_gasto_ron)/SUM(factor) AS media_ron,
    SUM(factor_gasto_tequila)/SUM(factor) AS media_tequila,
    SUM(factor_gasto_vodka)/SUM(factor) AS media_vodka
  FROM consolidado_gh
) AS medias;
"""

varianzas = spark.sql(query) # Corre el query
````

````python
varianzas.show()
````

````python
varianzas.createOrReplaceTempView('vars')  # Añadimos nuestro dataset de consolidado al catálogo como consolidado_gh
````

### Covarianzas

````python
#Generamos un query para obtener las covarianzas de los productos de interes teniendo en cuenta el factor de expansión
query="""
SELECT
  SUM(factor*((A220-media_refresco)*(A215-media_agua)))/SUM(factor) AS cvar_refresco_agua,
  SUM(factor*((A220-media_refresco)*(A106-media_papas)))/SUM(factor) AS cvar_refresco_papas,
  SUM(factor*((A220-media_refresco)*(A209-media_golosinas)))/SUM(factor) AS cvar_refresco_golosinas,
  SUM(factor*((A220-media_refresco)*(A230-media_ron)))/SUM(factor) AS cvar_refresco_ron,
  SUM(factor*((A220-media_refresco)*(A233-media_tequila)))/SUM(factor) AS cvar_refresco_tequila,
  SUM(factor*((A220-media_refresco)*(A235-media_vodka)))/SUM(factor) AS cvar_refresco_vodka
FROM consolidado_gh
JOIN (
  SELECT
    SUM(factor_gasto_refresco)/SUM(factor) AS media_refresco,
    SUM(factor_gasto_agua)/SUM(factor) AS media_agua,
    SUM(factor_gasto_papas)/SUM(factor) AS media_papas,
    SUM(factor_gasto_golosinas)/SUM(factor) AS media_golosinas,
    SUM(factor_gasto_ron)/SUM(factor) AS media_ron,
    SUM(factor_gasto_tequila)/SUM(factor) AS media_tequila,
    SUM(factor_gasto_vodka)/SUM(factor) AS media_vodka
  FROM consolidado_gh
) AS medias;
"""

covarianzas = spark.sql(query) # Corre el query
````

````python
covarianzas.show()
````

````python
covarianzas.createOrReplaceTempView('cvars')  # Añadimos nuestro dataset de consolidado al catálogo como consolidado_gh
````

### Desviación estandard

````python
#Generamos un query para obtener las desviaciones estandar de los productos de interes teniendo en cuenta el factor de expansión
query="""
SELECT
  SQRT(var_refresco) AS desvest_refresco,
  SQRT(var_agua) AS desvest_agua,
  SQRT(var_papas) AS desvest_papas,
  SQRT(var_golosinas) AS desvest_golosinas,
  SQRT(var_ron) AS desvest_ron,
  SQRT(var_tequila) AS desvest_tequila,
  SQRT(var_vodka) AS desvest_vodka
FROM vars;
"""

desv_est = spark.sql(query) # Corre el query
````

````python
desv_est.show()
````

````python
desv_est.createOrReplaceTempView('desv_ests')  # Añadimos nuestro dataset de consolidado al catálogo como consolidado_gh
````

### Correlaciones

````python
#Generamos un query para obtener la correlacion de los productos de interes teniendo en cuenta el factor de expansión
query="""
SELECT
  cvars.cvar_refresco_agua/(desvest_refresco*desvest_agua) AS correl_refresco_agua,
  cvar_refresco_papas/(desvest_refresco*desvest_papas) AS correl_refresco_papas,
  cvar_refresco_golosinas/(desvest_refresco*desvest_golosinas) AS correl_refresco_golosinas,
  cvar_refresco_ron/(desvest_refresco*desvest_ron) AS correl_refresco_ron,
  cvar_refresco_tequila/(desvest_refresco*desvest_tequila) AS correl_refresco_tequila,
  cvar_refresco_vodka/(desvest_refresco*desvest_vodka) AS correl_refresco_vodka
FROM cvars, desv_ests
"""

correlaciones = spark.sql(query) # Corre el query
````

````python
correlaciones.show()
````

````python
#Generamos un query para resumir el valor del gasto en cada producto por estado
query="""
SELECT
  estado AS CODIGO,
  CAST(SUM(factor_gasto_refresco) AS numeric(20,2)) AS consumo_refresco,
  CAST(SUM(factor_gasto_agua) AS numeric(20,2)) AS consumo_agua,
  CAST(SUM(factor_gasto_papas) AS numeric(20,2)) AS consumo_papas,
  CAST(SUM(factor_gasto_golosinas) AS numeric(20,2)) AS consumo_golosinas,
  CAST(SUM(factor_gasto_ron) AS numeric(20,2)) AS consumo_ron,
  CAST(SUM(factor_gasto_tequila) AS numeric(20,2)) AS consumo_tequila,
  CAST(SUM(factor_gasto_vodka) AS numeric(20,2)) AS consumo_vodka
FROM consolidado_gh
GROUP BY estado
ORDER BY CODIGO;
"""

consumo_estado = spark.sql(query) # Corre el query
````

````python
consumo_estado.show()
````

````python
consumo_estado = consumo_estado.toPandas()
````

````python
columnas = ['consumo_refresco', 'consumo_agua', 'consumo_papas', 'consumo_golosinas', 'consumo_ron', 'consumo_tequila', 'consumo_vodka']
for i in columnas:
  consumo_estado[i]=consumo_estado[i].astype(float)
````

````python
shapefile = pd.merge(
    left=shapefile,
    right=consumo_estado,
    on='CODIGO',
    how='left'
)
````

````python
consumo=consumo_estado.consumo_refresco.to_list()
linear = cm.LinearColormap(['green','yellow','red'],vmin=min(consumo), vmax=max(consumo))
````
